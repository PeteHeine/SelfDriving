Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        #print(line)
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(initial_img,img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)


def drawLine_HoughValues(img,thetaMean,rhoMean,lineColor):
    newImg = img.copy()
    for iLine in range(0,len(rhoMean)):
        a = np.cos(thetaMean[iLine]-math.pi/2) # with "-math.pi/2" the theta values are correct to hough space
        b = np.sin(thetaMean[iLine]-math.pi/2) # with "-math.pi/2" the theta values are correct to hough space
        x0 = a*rhoMean[iLine]
        y0 = b*rhoMean[iLine]
        
        x1 = int(x0 + 2000*(-b))   # Here i have used int() instead of rounding the decimal value, so 3.8 --> 3
        y1 = int(y0 + 2000*(a))    # But if you want to round the number, then use np.around() function, then 3.8 --> 4.0
        x2 = int(x0 - 2000*(-b))   # But we need integers, so use int() function after that, ie int(np.around(x))
        y2 = int(y0 - 2000*(a))
        
        cv2.line(newImg,(x1,y1),(x2,y2),lineColor,2)
    return newImg

def myDrawLine(img,thetaMean,rhoMean,lineColor):
    y1 = int(img.shape[0])
    y2 = int(img.shape[0]*3/5)
    newImg = img.copy()
    for iLine in range(0,len(rhoMean)):
        a = np.cos(thetaMean[iLine]-math.pi/2) # with "-math.pi/2" the theta values are correct to hough space
        b = np.sin(thetaMean[iLine]-math.pi/2) # with "-math.pi/2" the theta values are correct to hough space
        x0 = a*rhoMean[iLine]
        y0 = b*rhoMean[iLine]
        
        alpha = np.tan(thetaMean[iLine])
        beta = y0-alpha*x0
        x1 = int((y1-beta)/alpha)
        x2 = int((y2-beta)/alpha)
        
        cv2.line(newImg,(x1,y1),(x2,y2),lineColor,2)
    return newImg
```

```python
import os
dirsImage = os.listdir("test_images/")

for x in range(0,len(dirsImage)):
    print(dirsImage[x])

#for x in range(0,1):
dirImage = dirsImage[5]
#img = cv2.imread("test_images/" + dirImage)
rgb = mpimg.imread("test_images/" + dirImage)


def detectAndFilterLines(rgb):
    # Convert rgb to gray
    img = grayscale(rgb)
    
    # Blur image
    kernel_size = 5
    imgBlur = gaussian_blur(img, kernel_size)

    # Detect edges
    lThreshold = 100
    hThreshold = 150
    imgEdge = canny(imgBlur, lThreshold, hThreshold)

    # Mask image
    topPositionRatio = 6.5/10
    fromCenterSpreadingRatio = 1/20

    dimImage = img.shape
    vertices = np.array([[(0,dimImage[0]),(dimImage[1]/2-dimImage[1]*fromCenterSpreadingRatio,dimImage[0]*topPositionRatio),(dimImage[1]/2+dimImage[1]*fromCenterSpreadingRatio,dimImage[0]*topPositionRatio),(dimImage[1],dimImage[0])]], dtype=np.int32)
    imgEdgeMasked = region_of_interest(imgEdge, vertices)

    ## Detect lines segments in hough space
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/45 # angular resolution in radians of the Hough grid
    threshold = 30     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 #minimum number of pixels making up a line
    max_line_gap =10    # maximum gap in pixels between connectable line segments
    
    lines = cv2.HoughLinesP(imgEdgeMasked, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)
    edgeColor = [255,0,0]
    edgeColor2 = [0,0,255]
    edgeThickness = 2
    cLines = 0

    
    ## Lines are sorted in angle histogram bins. 
    # Each bin collects lines with similar angle...
    # Important feature: Each line is weighted by its length
    
    # Angle intervals
    angleBinsInterval =10*math.pi/180
    # Angle spread
    angleSpreadMinMax = (-90*math.pi/180 , 90*math.pi/180)
    angleSpread = np.diff(angleSpreadMinMax);
    # Number of bins
    nAngleBinsInterval = np.round(angleSpread/angleBinsInterval)
    # An counter foreach bin in the "angle histogram"
    cAngleBins = np.zeros(int(nAngleBinsInterval))
    lineLength = np.zeros(len(lines))
    lineTheta = np.zeros(len(lines))
    lineRho = np.zeros(len(lines))
    lineRhoTest = np.zeros(len(lines))
    
    for line in lines:    
        for x1,y1,x2,y2 in line:
        
            lineLength[cLines] = np.sqrt(pow(x2-x1,2)+pow(y2-y1,2))
            lineTheta[cLines] = np.arctan((y2-y1)/(x2-x1)) # Subtract with pi/2 to get the hough definition os theta
            lineRho[cLines] = x1*np.cos(lineTheta[cLines]-math.pi/2)+y1*np.sin(lineTheta[cLines]-math.pi/2)
            #lineRhoTest[cLines] = x2*np.cos(lineTheta[cLines]-math.pi/2)+y2*np.sin(lineTheta[cLines]-math.pi/2)
        
            # To visualize 
            #cv2.line(imgLines2, (x1, y1), (x2, y2), edgeColor, edgeThickness)
            #cv2.putText(imgLines2, str(cLines), (int((x2-x1)/2+x1), int((y2-y1)/2+y1)),cv2.FONT_HERSHEY_SIMPLEX,1,255,3)
            #print(str(cLines), " x1 y1 x2 y2: ",line, ", Length: ", lineLength[cLines], ", Angle (theta): ",lineTheta[cLines])
        
            # Each line is store in a "angle" bin with a weight (the length of the line)
            cIdx = math.floor((lineTheta[cLines]-angleSpreadMinMax[0])/angleBinsInterval)
            cAngleBins[cIdx] = cAngleBins[cIdx]+lineLength[cLines]
            cLines = cLines+1
    
    # AT THIS POINT CONSIDER FILTERING OUT SPECIFIC ANGLES IN THE ANGLE HISTOGRAM (SET THEM TO 0)
    selectAngle = 30 # (deg)
    filterAngle = (-selectAngle*math.pi/180,selectAngle*math.pi/180)

    binIdx = np.arange(0.0,nAngleBinsInterval,1)
    centerBinValues = binIdx*angleBinsInterval+angleBinsInterval/2+angleSpreadMinMax[0]
    turnOfBins = (centerBinValues>filterAngle[0]) & (centerBinValues<filterAngle[1])
    #print(turnOfBins)
    cAngleBins[turnOfBins] = 0
    
    ## Small script for peak detection in the histogram
    # Filter indexes
    cAngleBinsFiltered = cAngleBins
    cAngleBinsFiltered = np.convolve(cAngleBins,[0.25, 0.5, 1, 0.5, 0.25],'same') 

    
    ## Angle bins with most incidences are selected first.
    # An angle bin is only selected if neighbors are smaller 
    # The algorithm will stop iterating when two max bins have been selected.
    
    
    # Index values of the sorted array
    maxIdx = np.flipud(cAngleBinsFiltered.argsort())
    #maxIdx = np.delete(maxIdx,cAngleBinsFiltered[maxIdx]==0)
    
    maxIdxFiltered = np.array([])
    for iIdx in maxIdx:
        idxTop = min(iIdx+1,len(cAngleBinsFiltered)-1)
        idxDown = max(iIdx-1,0)
        peakState = (cAngleBinsFiltered[iIdx]>=cAngleBinsFiltered[idxDown] and cAngleBinsFiltered[iIdx]>=cAngleBinsFiltered[idxTop])
        #print(idxTop,idxDown,peakState)

        if(peakState):
            maxIdxFiltered = np.append(maxIdxFiltered,iIdx)
        
        if(len(maxIdxFiltered)>=2):
            break;

    ## All lines with the 
    minInterval = maxIdxFiltered*angleBinsInterval+angleSpreadMinMax[0]
    maxInterval = minInterval+angleBinsInterval
    lineIdx = np.zeros((len(lineTheta),2),dtype=bool)
    thetaMean = []
    rhoMean = []
    lineIdx[:,0] = (lineTheta>=minInterval[0]) & (lineTheta<=maxInterval[0])
    lineIdx[:,1] = (lineTheta>=minInterval[1]) & (lineTheta<=maxInterval[1])
    
    for iLine in range(0,len(maxIdxFiltered)):
        #if(False):
        #    print("iLine",iLine)
        #    print("minInterval",minInterval,"maxInterval",maxInterval)
        #    print("lineTheta",lineTheta)
        #    print("cAngleBins",cAngleBins)
        #    print("lineIdx[:,iLine]",lineIdx[:,iLine])
        #    print("lineTheta[lineIdx[:,iLine]]",lineTheta[lineIdx[:,iLine]])
        #    print("lineLength[lineIdx[:,iLine]]",lineLength[lineIdx[:,iLine]])
        #    print("lineRho[lineIdx[:,iLine]]",lineRho[lineIdx[:,iLine]])
        if(lineTheta[lineIdx[:,iLine]].size>0):
            thetaMean = np.append(thetaMean,np.average(lineTheta[lineIdx[:,iLine]], weights=lineLength[lineIdx[:,iLine]]))
            rhoMean = np.append(rhoMean,np.average(lineRho[lineIdx[:,iLine]], weights=lineLength[lineIdx[:,iLine]]))
        #thetaMean[iLine] = np.average(lineTheta[lineIdx[:,iLine]], weights=lineLength[lineIdx[:,iLine]])



    
    return thetaMean, rhoMean

thetaMean, rhoMean = detectAndFilterLines(rgb)


newImg = myDrawLine(rgb,thetaMean,rhoMean,(0,255,0))
#newImg = drawLine_HoughValues(rgb,thetaMean,rhoMean,(0,255,0))
plt.imshow(newImg,cmap='gray')


    
```

### MyReflecions
I have implemented few tricks to (hopefully) improve lane detection. A histogram of angles, where each angle/line is weighted by its length (this it NOT the rho value), thus making longer lines "count"/weigh more than small lines. The histogram is filtered first with a mask to remove horisontal lines and then a convolution to smooth out the histogram to be more robust to small glitches. A peak detector runs through angle bins to detect angles with the highest incidence. The peak detector will only detect a peak if its neighbors have lower incidence. The detector runs through highest incidence angles until two peak are determined (the peak detector may only detect one and only draw one). 
Possible improvements:
- Operate, detect lanes and filter directly in hough space
- Camera information (camera height, angling, FOV) can be used to project the image "unto the road" (inverse perspecitive mapping). Lanes would then create parallel lines (same theta), which are more visible in the hough space. 
- Use information from previous frames under the assumption that lanes will not jump or move significantly from frame to frame. Track lines or line parameters using kalman filter or maybe add the hough space of the previous frame to the current frame. In the above code I would remember/add the angle histogram of the previous to the new angle histogram.
- And alot of parameter tuning.
- Potentially adding some kind of spline to detect curved lines.
